```python
import pandas as pd
```

```python
df0 = pd.read_csv('rightmove_data-2.csv')
```

```python
df0.head(5)
```

```python
df = df0.drop('Links',axis=1)
```

```python
df.reset_index(inplace=True, drop=True)
```

```python
#Price Column to numeric type
df["Price"].replace(regex=True, inplace=True, to_replace=r"\D", value=r"")
df["Price"] = pd.to_numeric(df["Price"])
```

```python
#Get short postcode from Address
pat = r"\b([A-Za-z][A-Za-z]?[0-9][0-9]?[A-Za-z]?)\b"
df["Postcode"] = df["Address"].astype(str).str.extract(pat, expand=True)[0]
```

```python
#Get Full_Postcode from Address
pat = r"([A-Za-z][A-Za-z]?[0-9][0-9]?[A-Za-z]?[0-9]?\s[0-9]?[A-Za-z][A-Za-z])"
df["Full_Postcode"] = df["Address"].astype(str).str.extract(pat, expand=True)[0]
```

```python
 # Extract number of bedrooms from Description to a separate column
 pat = r"\b([\d][\d]?)\b"
 df["number_bedrooms"] = df["Description"].astype(str).str.extract(pat, expand=True)[0]
 df.loc[df["Description"].str.contains("studio", case=False), "number_bedrooms"] = 0
 df["number_bedrooms"] = pd.to_numeric(df["number_bedrooms"])
```

```python
# Clean white spaces and newlines in Description column
df["Description"] = df["Description"].str.strip("\n").str.strip()
```
