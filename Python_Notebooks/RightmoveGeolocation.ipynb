```python
# load dataset
df = pd.read_csv("rmclean_29-12.csv")
```

```python
#Using the google API to get coordinates
def get_coordinates(address):
    geocode_result = gmaps.geocode(str(address))
    if len(geocode_result) > 0:
        return list(geocode_result[0]['geometry']['location'].values())
    else:
        return [np.NaN, np.NaN]
```

```python

coordinates2 = df['Address'].apply(lambda x: pd.Series(get_coordinates(x), index=['Latitude', 'Longitude']))
```
